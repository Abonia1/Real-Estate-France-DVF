## <center> Immobilier Data Pre-processing</center>
Datset is given in text format so we will convert the file format to csv and then import it in Oracle MySQL in order to load to do data analysis for later.

As data of DVF are provided based on year,to do a data analysis we will concate the last five year data file into single file.

```python
import csv
import pandas as pd

txt_list=['Dataset/DVF2019.txt','Dataset/DVF2018.txt','Dataset/DVF2017.txt','Dataset/DVF2016.txt','Dataset/DVF2015.txt']
csv_list=['Dataset/DVF2019.csv','Dataset/DVF2018.csv','Dataset/DVF2017.csv','Dataset/DVF2016.csv','Dataset/DVF2015.csv']
for txtpath,csvpath in zip(txt_list, csv_list):
    with open(txtpath, 'r') as in_file:
        stripped = (line.strip() for line in in_file)
        lines = (line.split("|") for line in stripped if line)
        with open(csvpath, 'w') as out_file:
            writer = csv.writer(out_file)
            #writer.writerow(('Code service CH','Reference document','1 Articles CGI','2 Articles CGI','3 Articles CGI','4 Articles CGI','5 Articles CGI','No disposition','Date mutation','Nature mutation','Valeur fonciere','No voie','B/T/Q','Type de voie','Code voie','Voie','Code postal','Commune','Code departement','Code commune','Prefixe de section','Section','No plan','No Volume','1er lot','Surface Carrez du 1er lot','2eme lot','Surface Carrez du 2eme lot','3eme lot','Surface Carrez du 3eme lot','4eme lot','Surface Carrez du 4eme lot','5eme lot','Surface Carrez du 5eme lot','Nombre de lots','Code type local','Type local','Identifiant local','Surface reelle bati','Nombre pieces principales','Nature culture','Nature culture speciale','Surface terrain'))
            writer.writerows(lines)
```
